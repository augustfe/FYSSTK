```python
from grad_analysis import GradAnalysis
import numpy as np
```

```python
def simple_poly(x: float) -> float:
    return 4 - 3 * x + 2 * x**2
```

```python
np.random.seed(2023)
n_points = 100
x = np.random.randn(n_points)
y = simple_poly(x) + np.random.randn(n_points)

x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
true_theta = np.array([4, -3, 2]).reshape(-1, 1)
n_epochs = 500
```

```python
GA = GradAnalysis(
    x,
    y,
    n_epochs=n_epochs,
    seed=2023,
    model="OLS",
    method="analytic",
    true_theta=true_theta,
    target_func=simple_poly,
)
```

```python
from Schedules import Constant

eta_num = 75
eta_arr = np.logspace(-5, -1, eta_num)
GA.constant_analysis(eta_arr)
```

```python
eta_num = 75
rho_num = 75

eta_arr = np.logspace(-5, -1, eta_num)
rho_arr = np.linspace(1 / n_points, 1, rho_num)

GA.momentum_analysis(eta_arr, rho_arr)
```

```python
# NOTE: Adagrad seems to be less sensitive to the learning rate

eta_arr = np.logspace(-3, 0, eta_num)
GA.adagrad_analysis(eta_arr)
```

```python
eta_arr = np.logspace(-5, 0, eta_num)
rho_arr = np.linspace(1 / n_points, 0.99, rho_num)

GA.adagrad_momentum_analysis(eta_arr, rho_arr)
```

```python
rho2_num = 75
rho2_arr = np.linspace(0.05, 0.999, rho2_num)

GA.adam_analysis(eta_arr, rho_arr, rho2_arr)
```
