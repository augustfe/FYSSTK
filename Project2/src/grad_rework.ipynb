```python
from grad_analysis import GradAnalysis
import numpy as np
```

```python
def simple_poly(x: float) -> float:
    return 4 - 3 * x + 2 * x ** 2
```

```python
np.random.seed(2023)
n_points = 100
x = np.random.randn(n_points)
y = simple_poly(x) + np.random.randn(n_points)
base_theta = np.array([4, -3, 2])
n_epochs = 500
```

```python
GA = GradAnalysis(x, y, n_epochs=n_epochs, seed=2023, model="OLS", method="analytic", true_theta=base_theta, target_func=simple_poly)
```
